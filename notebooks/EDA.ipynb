# 游늵 EDA - Exploratory Data Analysis

Este notebook contiene el an치lisis exploratorio del dataset `gastos.csv`.

```
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('../data/gastos.csv')
df['fecha'] = pd.to_datetime(df['fecha'])
df.head()
```

## Limpieza de Datos

```
df.info()
```

```
# Verificar nulos
df.isnull().sum()
```

## Estad칤sticas Descriptivas

```
df.describe()
```

## Total por usuario

```
df.groupby('usuario')['monto'].sum()
```

## Total por categor칤a

```
df.groupby('categoria')['monto'].sum()
```

## Gr치fico: Total por categor칤a

```
df.groupby('categoria')['monto'].sum().plot(kind='bar'); plt.show()
```

## Gr치fico: Evoluci칩n diaria
