# EDA - Exploratory Data Analysis

Este notebook contiene el análisis exploratorio del dataset `gastos.csv`.

```
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('/data/gastos.csv')
df['fecha'] = pd.to_datetime(df['fecha'])
df.head()
```

## Limpieza de Datos

```
df.info()
```

```
# Verificar nulos
df.isnull().sum()
```

## Estadísticas Descriptivas

```
df.describe()
```

## Total por usuario

```
df.groupby('usuario')['monto'].sum()
```

## Total por categoría

```
df.groupby('categoria')['monto'].sum()
```

## Gráfico: Total por categoría

```
df.groupby('categoria')['monto'].sum().plot(kind='bar'); plt.show()
```

## Gráfico: Evolución diaria

```
df.groupby('fecha')['monto'].sum().plot(figsize=(10,4)); plt.show()
```
